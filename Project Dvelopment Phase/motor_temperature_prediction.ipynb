```python
df = df.sample(n=60000, random_state=42)
df.shape
```

```python
X = df.drop('pm', axis=1)
y = df['pm']
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
X_scaled = mm.fit_transform(X)

X = pd.DataFrame(X_scaled, columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=3
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

lr = LinearRegression()
dr = DecisionTreeRegressor()
rf = RandomForestRegressor()
```

```python
lr.fit(X_train, y_train)
dr.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
p1 = lr.predict(X_test)
p2 = dr.predict(X_test)
p3 = rf.predict(X_test)
```

```python
from sklearn import metrics

print("Linear Regression:", metrics.r2_score(y_test, p1))
print("Decision Tree:", metrics.r2_score(y_test, p2))
print("Random Forest:", metrics.r2_score(y_test, p3))
```
